<p id='uni' style='padding:15px;background-color:#4a235a;text-align:center;font-size:30px;font-weight:800;border-radius:20px;color:#2efbcd'> Univariate Plotting with Pandas

```python
# Importamos las librerias 
import pandas as pd
```

```python
# Cargamos el DataFrame
reviews = pd.read_csv('../datasets/winemag-data_first150k.csv',index_col=0)

# Revisamos que el DataFrame haya cargado correctamente
reviews.head(3)
```

<p id='uni_1' style='text-align:center;background-color:#f4c272;color:#008b8b;font-weight:800;font-size:25px;border-radius:15px;padding:10px;'>Bar Charts and Categorical Data</p>

El <b style='color:#2efbcd'>Gráfico de Barras</b> es el más común y simple de todos, muestran categorías como numeros, por ejemplo:
- La cantidad de huevos consumidos en el desayuno (una categoría).

O en nuestro caso:
- Producción de Vino en distintas provincias del mundo.

```python
reviews['province'].value_counts().head(10).plot.bar(color=['blue','orange','green','red','magenta','brown','pink','grey','yellow','cyan']);
```

Con el gráfico podemos deducir que <b style='color:#44BEE4'>California</b> produce con gran diferencia más <b style='color:brown'>vino</b> que los demás. Deberíamos preguntarnos cuál es el porcentaje que toma, porque este gráfico no nos dice nada más que números.

> <b style='font-size:20px'>Consejo:</b>
>
> - dataframe.shape = [Filas, Columnas]
>
> - dataframe.size = [Filas * Columnas]

```python
(reviews['province'].value_counts().head(10) / reviews['province'].size).plot.bar(color=['blue','orange','green','red','magenta','brown','pink','grey','yellow','cyan']);
```

Ahora sabemos que <b style='color:#44bee4'>California</b> produce casi el <b style='color:#44bee4'>30%</b> de todo el vino producido en el mundo.

> <b style='font-size:20px'>Consejo:</b>
>
> Los <b style='color:#2efbcd'>Gráficos de Barra</b> son muy utiles cuando quieres representar la <b style='color:#44bee4'>altura</b> con algún <b style='color:#44bee4'>número</b>; De igual forma cada <b style='color:#ffcc00'>columna</b> puede representar lo que sea mientras sea <b style='color:#ffcc00'>categorico</b>.

En valores <b style='color:#44bee4'>nominales</b> (como este caso), no es necesario ordenar los valores. Pero en casos <b style='color:#ffcc00'>ordinales</b> sí es necesario para mejor entendimiento.

Por ejemplo:

```python
reviews['points'].value_counts().sort_index().plot.bar(color=['blue','orange','green','red','purple','brown','pink','grey','yellow','cyan','darkblue']);
```
